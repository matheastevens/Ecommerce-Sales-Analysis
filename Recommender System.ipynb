```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity
```

```python
df = pd.read_csv('Data/cleaned_sales_data2')
df.dropna(inplace = True)
df
```

```python
pivot = pd.pivot_table(df, index = 'StockCode', columns = ['Customer ID', "Country"], values = 'Quantity')

recommender = pd.DataFrame(cosine_similarity(sparse.csr_matrix(pivot.fillna(0))), columns = pivot.index, 
                              index = pivot.index)
recommender.head()
```
