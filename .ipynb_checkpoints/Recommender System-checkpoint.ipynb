## Recommender System Build

This recommender system is immplemented in the Streamlit web application. It was developed in this notebook for proof of concept. 

```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity
```

```python
df = pd.read_csv('Data/cleaned_sales_data2')
df.dropna(inplace = True)                           #In the cleaned dataset, we retained some NaN's in Customer ID.
df                                                  #because Customer ID is a main feature in establishing a recommendation
                                                    #system, we will drop all NaN's for this model.
```

```python
#Building this recommender system based on Customer ID and Country. It is an item-based Collaborative Recommender. 
pivot = pd.pivot_table(df, index = 'StockCode', columns = ['Customer ID', "Country"], values = 'Quantity')

recommender = pd.DataFrame(cosine_similarity(sparse.csr_matrix(pivot.fillna(0))), columns = pivot.index, 
                              index = pivot.index)             #Using the cosine similarity to establish similar products. 
recommender.head()
```
